```
filename = "https://raw.githubusercontent.com/wcota/covid19br/master/cases-brazil-states.csv"
df = pd.read_csv(filename)
```

```
['epi_week','country','newDeaths','deaths','newCases',	'totalCases',	'deathsMS',	'totalCasesMS',	'deaths_per_100k_inhabitants',	'totalCases_per_100k_inhabitants',	'deaths_by_totalCases',	'recovered','suspects','tests','tests_per_100k_inhabitants']
```

#Cleaning SP

```
df_sp = df[df['state'] == 'SP']
df_sp.head()
df_sp.groupby(['date']).sum()
df_sp_newdeath = df_sp.drop(columns=['epi_week','country','newDeaths','deaths',	'newCases',	'totalCases',	'deathsMS',	'totalCasesMS',	'deaths_per_100k_inhabitants',	'totalCases_per_100k_inhabitants',	'deaths_by_totalCases','recovered','tests','tests_per_100k_inhabitants'])
df_sp_transposed = df_sp_newdeath.transpose()
df_sp_clean = df_sp_transposed.drop(index = ['state','city'])
df_sp_clean.columns = df_sp_clean.iloc[0]
df_sp_clean = df_sp_clean.drop('date')
df_sp_clean['state'] = 'SP'
df_sp_clean
```

```
df_total = pd.DataFrame()
def select_columns(data_frame, column_names):
    new_frame = data_frame.loc[:, column_names]
    return new_frame
```

```
df_total = select_columns(df_sp_clean,df_sp_clean.columns)
```

# Appending States

```
for estado in df['state'].unique():
  df_al = df[df['state'] == estado]
  df_al.groupby(['date']).sum()
  df_al_newdeath = df_al.drop(columns=['epi_week','country','newDeaths','newCases',	'deaths',	'newCases',	'deathsMS',	'totalCasesMS',	'deaths_per_100k_inhabitants',	'totalCases_per_100k_inhabitants',	'deaths_by_totalCases','recovered','tests','tests_per_100k_inhabitants'])
  df_al_transposed = df_al_newdeath.transpose()
  df_al_clean = df_al_transposed.drop(index = ['state','city'])
  df_al_clean.columns = df_al_clean.iloc[0]
  df_al_clean = df_al_clean.drop('date')
  df_al_clean['state'] = estado
  df_total = df_total.append(df_al_clean)
```

#Dropping duplicates

```
df_total.sort_values('state', inplace= True)
df_total.drop_duplicates(subset ="state", keep = 'first', inplace = True) 
```
